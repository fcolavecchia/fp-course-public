# Dictionaries

A `Map` is a dictionary, so the data is accessed through a key.

```python
let m1 = Map [3, "Apples" ; 2, "Bananas"]

printfn "%A" m1 
printfn "%A" m1[3] 
```

```python
printfn "%A" m1.[5]
```

Unlike lists, it is possible to create new dictionaries using functions to add or change certain values: The method `Map.add` adds a (key,value) pair if the key is not present in the dictionary, or replaces the value if the key is already present. In both cases, it returns a new map.

```python
let m2 = m1.Add(4,"Oranges")
printfn "%A" m2
```

```python
let m3 = Map.add 2 "Peaches" m2 
printfn "%A" m2
printfn "%A" m3
```

```python
let toStringPlayer number player = 
    $"{number}: {player.Name} ({player.Age}), {player.Position}, juega en {player.Team}"


let prettyPrintMap m = 
    m
    |> Map.iter (fun n p -> printfn "%s" (toStringPlayer n p))  

prettyPrintMap champions2022       
```

The `Map` type presents a reduced list of methods, and sometimes one needs to transform it into a list. For example, if we want to sort the map by the keys, one transforms the dictionary into a list of tuples `(key,value)` with `Map.toList`, use a `list` method and returns to the `map` with `Map.ofList`: 

```python
// Sort by key
champions2022 
|> Map.toList 
|> List.sortBy (fun (n,p) -> n)
|> Map.ofList
|> prettyPrintMap
```

Perhaps one does not need to print all the data for each player:

```python
champions2022 
|> Map.map (fun _ p -> $"{p.Name} tiene {p.Age} años")
|> Map.iter (fun _ s -> printfn "%s" s)
```

Any element of the dictionary can be searched through the key:

```python
champions2022
|> Map.find 22
```

```python
champions2022
|> Map.find 31
```

Note that `Map.find` obviously returns the only element it finds. If one wants all elements that share a certain characteristic, we use `Map.filter`

```python
champions2022
|> Map.filter (fun _ p -> p.Age = 21uy)
```

```python
let arqueros = 
    champions2022
    |> Map.filter (fun _ p -> p.Position = GoalKeeper)

prettyPrintMap arqueros    
```

We can use other methods to process the data from the dictionary, but as before, it may be necessary to convert them to a list with `Map.ofList`:

```python
champions2022
|> Map.toList 
|> List.averageBy (fun (_,p) -> float p.Age)
```

The `Map.partition` function allows us to group the elements into two dictionaries with disjoint characteristics:

```python
let youngerThan30, olderOnes = 
    champions2022
    |> Map.map (fun _ p -> (p.Name,p.Age))
    |> Map.partition (fun _ (name,age) -> age < 30uy)

printfn "There are %A players younger than 30 and " (youngerThan30.Count)    
printfn "%A players older  " (olderOnes.Count) 
```

Finally, some functions allow us to evaluate a certain characteristic of a dictionary:

```python
champions2022
|> Map.forall (fun _ p -> p.Age < 30uy)
```
